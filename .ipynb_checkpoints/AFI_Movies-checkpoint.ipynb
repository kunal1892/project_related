```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/AFI%27s_100_Years...100_Movies"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}
page=requests.get(url,headers=headers)
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
tdata=soup.find(class_="wikitable sortable sticky-header")
tdata
```

```python
headings=[h.get_text().strip() for h in tdata.find_all('th')]
headings
```

```python
data=[[t.get_text().strip() for t in d.find_all('td')] for d in tdata.find_all('tr')[1:]]
data
```

```python
d={k:[] for k in headings}
for p in data:
    d['Film'].append(p[0])
    d['Release year'].append(p[1])
    d['Director'].append(p[2])
    d['Production companies'].append(p[3])
    d['1998 Rank'].append(p[4])
    d['2007 Rank'].append(p[5])
d    
```

```python
df=pd.DataFrame(d)
df
```
