```python
data=union_territories_data.find_all('tr')[1:]
data
```

```python
result=[]
for f in data:
    res=[t.get_text().strip() for t in f.find_all('td')]
    res=[t[:t.index('[')] if '[' in t else t for t in res]
    result.append(res)
result 
```

```python
d = {k:[] for k in headings}

for q in result[:-1]:

    d['ISO'].append(q[0])

    d['Vehicle  code'].append(q[1])

    d['Zone'].append(q[2])

    d['Capital'].append(q[3])

    d['Largest city'].append(q[4])

    d['Established'].append(q[5])

    d['Population  (2011)'].append(q[6])

    d['Area  (km2)'].append(q[7])

    d['Official  languages'].append(q[8])

    d['Additional official  languages'].append(q[9])
d['union_territories']=[]
d['union_territories'].extend(union_territories)
```
