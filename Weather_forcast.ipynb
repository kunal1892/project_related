```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://forecast.weather.gov/MapClick.php?lat=37.7801&lon=-122.4202"
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
all_data=soup.find_all(class_ ="tombstone-container")
all_data
```

```python
first=all_data[0]
first
```

```python
pn=first.find(class_="period-name").get_text()
pn
```

```python
tp=first.find(class_="temp temp-low").get_text()
tp
```

```python
sd=first.find(class_="short-desc").get_text()
sd
```

```python
pns,temp,sds=([],[],[])
for f in all_data:
    pns.append(f.find(class_="period-name").get_text())
    temp.append(f.find(class_="temp").get_text())
    sds.append(f.find(class_="short-desc").get_text())
len(pns), len(temp), len(sds)    
    
```

```python
d={"period name":pns,"Temprature":temp,"Short Desc":sds}
d
```

```python
df=pd.DataFrame(d)
df
```

```python
df.to_csv("Weather.csv",index=False)
```
