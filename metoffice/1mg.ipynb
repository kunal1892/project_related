```python
from bs4 import BeautifulSoup

import requests

import pandas as pd

import string

names, sizes, prices = [],[],[]

for alpha in string.ascii_lowercase:

    for i in range(1, 11):

        url=f"https://www.1mg.com/drugs-all-medicines?page={i}&label={alpha}"

        headers = {

            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'

        }

        page = requests.get(url, headers = headers)

        soup = BeautifulSoup(page.content, 'html5lib')

        data = soup.find_all(class_="curved grow shrink AllMedicinespage__cardContainer__tozds")

        for f in data:

            names.append(f.find(class_="bodyMediumBold").get_text())

            sizes.append(f.find(class_="truncateTo1").get_text())

            prices.append(f.find(class_="flex alignBase gap-4").get_text())
 
len(names), len(sizes), len(prices)   
 
```

```python
df=pd.DataFrame({"Names":names,"Sizes":sizes,"Prices":prices})
df
```
