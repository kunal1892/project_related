```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/International_Indian_Film_Academy_Awards"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}
page=requests.get(url,headers=headers)
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
all_data=soup.find(class_="sortable wikitable")
all_data
```

```python
headings=[h.get_text().strip() for h in all_data.find_all('th')]
headings
```

```python
tdata= all_data.find_all('tr')[1:]
tdata
```

```python
result=[]
for h in tdata:
    result.append([t.get_text().strip() for t in h.find_all('td')])
result    
```

```python
for t in result:
    if '21st' in t:
        t.append(t[-1])
        t.append(t[-1])
    else:
        if len(t) ==6:
            host,venue,city=t[3],t[4],t[5]
        elif len(t)==5:
            t.insert(3,host)
        elif len(t)==4:
            t.append(venue)
            t.append(city)
```

```python
for t in result:
    print(len(t))
```

```python
d={k:[] for k in headings}
for t in result:
    d['Ceremony'].append(t[0])
    d['Date'].append(t[1])
    d['Best Film'].append(t[2])
    d['Host(s)'].append(t[3])
    d['Venue'].append(t[4])
    d['City, Country'].append(t[5])
d
```

```python
df=pd.DataFrame(d)
df
```

```python
df.to_csv("IIFA_Awards.csv", index=False)
```
