```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://www.metoffice.gov.uk/research/climate/maps-and-data/uk-and-regional-series#Yearorder"
page=requests.get(url)
soup=BeautifulSoup(page.content,'html5lib')
soup
```

```python
regions = [op['value'] for op in soup.find(id="region").find_all('option')[1:]]
regions
```

```python
params = [op['value'] for op in soup.find(id="parameter").find_all('option')[1:]]
params
```

```python
import os
path='./metoffice'
for reg in regions:
    os.mkdir(f"{path}/{reg}")
    for pr in params:
        url2=f"https://www.metoffice.gov.uk/pub/data/weather/uk/climate/datasets/{pr}/ranked/{reg}.txt"
        page2=requests.get(url2)
        soup2=BeautifulSoup(page2.content,'lxml')
        with open(f"{path}/{reg}/{reg}_{pr}.txt",'w') as fobj:
            fobj.write(soup2.get_text())
```
