```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url="https://en.wikipedia.org/wiki/States_and_union_territories_of_India"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36'
}
page=requests.get(url,headers=headers)
soup=BeautifulSoup(page.content,'html.parser')
soup
```

```python
tdata=soup.find_all(class_="wikitable sortable plainrowheaders")
tdata
```

```python
len(tdata)
```

```python
state_data=tdata[0]
state_data
```

```python
union_teritorries_data=tdata[1]
union_teritorries_data
```

```python
heads=[h.get_text().strip() for h in state_data.find_all('th')[:11]]
headings=[t[:t.index('[')] if '[' in t else t for t in heads]
headings      
```

```python
States=[h.get_text().strip() for h in state_data.find_all('th')[11:]][:-1]
States
```

```python
data=state_data.find_all('tr')[1:]
data
```

```python
result=[]
for f in data:
    res=[t.get_text().strip() for t in f.find_all('td')]
    res=[t[:t.index('[')] if '[' in t else t for t in res]
    result.append(res)
result    
```

```python
for r in result:
    if len(r)==9:
        r.insert(3,r[3])
for r in result:
    print(len(r))
```

```python
d = {k:[] for k in headings}

for q in result[:-1]:

    d['ISO'].append(q[0])

    d['Vehiclecode'].append(q[1])

    d['Zone'].append(q[2])

    d['Capital'].append(q[3])

    d['Largest city'].append(q[4])

    d['Statehood'].append(q[5])

    d['Population(2011)'].append(q[6])

    d['Area(km2)'].append(q[7])

    d['Officiallanguages'].append(q[8])

    d['Additional officiallanguages'].append(q[9])

d['State'].extend(States)
 
```

```python
d
```

```python
df=pd.DataFrame(d)
df
```

```python
df.to_csv('States_data.csv',index=False)
```
